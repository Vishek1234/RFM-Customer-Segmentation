## Imoplementation (Code Execution)

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Load

```python
df= pd.read_csv("data_1.csv")
df.head()
```

## Data cleaning


```python
df.isnull().sum()

```

```python
df['Returns'].fillna(0,inplace=True)
```

## Date Conversion

```python
df['Purchase Date']= pd.to_datetime(df['Purchase Date'])
```

## Feature Engineering 

```python
df['Year']= df['Purchase Date'].dt.year
df['Month']= df['Purchase Date'].dt.month
df['Month_Name']= df['Purchase Date'].dt.month_name()
```

## Exploratory Data Analysis (EDA)

```python
df.groupby('Month_Name')['Total Purchase Amount'].sum().sort_values(ascending=False)
```

## Revenue Calculation

```python
df['Total Purchase Amount'].sum()
```

## ## Business Metrics Derivation
* **CLV** (*Customer Lifetime Value*)
* **AOV** (*Average Order Value*)
* **Orders** (*Total Orders*)



```python
clv= df.groupby('Customer ID')['Total Purchase Amount'].sum().reset_index()
clv.columns=['Customer ID','CLV']
aov= df.groupby('Customer ID')['Total Purchase Amount'].mean().reset_index()
aov.columns=['Customer ID','AOV']
orders= df.groupby('Customer ID').size().reset_index(name='Total Orders')
```

## RFM Analysis

```python
latest_date= df['Purchase Date'].max()
rfm= df.groupby('Customer ID').agg({'Purchase Date': lambda x :(latest_date-x.max()).days,
                                    'Customer ID':'count',
                                    'Total Purchase Amount':'sum'
                                   })
```

```python
rfm.columns=['Recency','Frequency','Monetary']
```

## RFM Scoring

```python
rfm['R_score']= pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_score']= pd.qcut(rfm['Frequency'],4,labels=[1,2,3,4])
rfm['M_score']= pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])
```

## RFM Score Aggregation

```python
rfm['RFM_Score']= rfm['R_score'].astype(str)+rfm['F_score'].astype(str)+rfm['M_score'].astype(str)
```

## Numeric conversion

```python
rfm['R_score']= pd.to_numeric(rfm['R_score'], errors = 'coerce')
rfm['F_score']= pd.to_numeric(rfm['F_score'], errors='coerce')
rfm['M_score']= pd.to_numeric(rfm['M_score'], errors='coerce')
```

## Segment Analysis

```python
def segment_customers (row):
    r= int(row['R_score'])
    f= int(row['F_score'])
    m= int(row['M_score'])
    if r == 4 and f == 4 and m == 4:
        return'Champions'
    elif r>= 3 and f>= 3 :
        return 'Loyal Customers'
    elif r<=2:
        return 'At Risk'
    elif f == 4:
        return 'Frequent Buyers'
    else:
        return 'Average Customers'
```

```python
rfm['Segment']=rfm.apply(segment_customers,axis=1)
rfm.head()
```

## Segment Count

```python
rfm['Segment'].value_counts()
```

## Segment Distribution

```python
plt.figure()
sns.countplot(x='Segment',data=rfm)
plt.xticks(rotation=45)
plt.title('Customer Segment Distribution')
plt.show()
```

## Revenue by Segment

```python
plt.figure()
rfm.groupby('Segment')['Monetary'].sum().plot(kind='bar')
plt.title('Revenu By Segment')
plt.xticks(rotation=45)
plt.show()
```

## Average Monetary Per Segment

```python
plt.figure()
rfm.groupby('Segment')['Monetary'].mean().plot(kind='bar')
plt.title('Average Spending per Segment')
plt.xticks(rotation=-45)
plt.show()
```

## Frequency Distribution

```python
plt.figure()
sns.histplot(rfm['Frequency'],bins=20)
plt.title('Frequency Distribution')
plt.show()
```
